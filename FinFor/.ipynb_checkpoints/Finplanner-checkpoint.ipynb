```python
def getResponse(url):
    response= requests.get(url)
    
    
    if(response.status_code==200):
        json_response = response.json()
        #jsonData = json.dumps(json_response, indent=4, sort_keys=True)
    else:
        print("Error receiving data", response.status_code)
    return json_response


```

```python
def get_value(json_input, lookup_key):
        value = list(json_iterator(json_input, lookup_key))
        val = value[0] if value else None
        #print(f'lookup_key -> value : {val}')
        return val
def json_iterator(json_input, lookup_key):
    if isinstance(json_input, dict):
        for k, v in json_input.items():
            #print(f'{k} -- {v}')
            if k == lookup_key:
                yield v
            else:
                yield from json_iterator(v, lookup_key)
    elif isinstance(json_input, list):
        for item in json_input:
            yield from json_iterator(item, lookup_key)
```

```python
BTC_price=get_value(btc_response,'USD')['price']
ETH_price=get_value(eth_response,'USD')['price']
```

```python
#Emergency Fund calculation
Emergency_fund=monthly_income*3
```

```python

def financial_health(portfolio_df,Emergency_fund):
    portfolio_fund=portfolio_df.sum().values[0]
    if portfolio_fund > Emergency_fund:
        print("Congratulations! You have enough money in your emergency fund.")
    elif portfolio_fund < Emergency_fund:
        print(f"Caution! You are short of reaching your emergency fund target. Gap is: ${Emergency_fund - portfolio_fund}")
    elif portfolio_fund == Emergency_fund:
        print("Congratulations! You have reached your financial target.")
```

```python
financial_health(portfolio_df,Emergency_fund)
```

#### Part 2 - Retirement Planning

```python
from MCForecastTools import MCSimulation
stock_basket=.40
bond_basket=.60
weights =[.40,.60]
num_years=30
```

```python
# Format current date as ISO format
NY = 'America/New_York'

edate = dt.date.today()
start=pd.Timestamp('2016-05-01', tz=NY).isoformat()
#print(start)
end=pd.Timestamp(edate, tz=NY).isoformat()
#print(end)
#end = pd.Timestamp('2016-10-16', tz='America/New_York').isoformat()
# Set the tickers
tickers = ["AGG", "SPY"]
# Set timeframe to '1D' for Alpaca API, 
timeframe = "1D"
# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
Five_year_data=alpaca_api.get_barset(tickers, timeframe, start=start, end=end, limit=1000).df
```

```python
Five_year_data.head()
```

```python
#Configure and execute a Monte Carlo Simulation of 500 runs and 30 years for the 40/60 portfolio.
MC_even_dist=MCSimulation(Five_year_data, weights=weights, num_simulation=1000, num_trading_days=252*num_years)
# Print the simulation input data
MC_even_dist.portfolio_data.head()
```

```python
# Run a Monte Carlo simulation to forecast five years cumulative returns
MC_even_dist.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
MC_even_dist.plot_distribution()
```

```python
# Fetch summary statistics from the Monte Carlo simulation results
# YOUR CODE HERE!
even_tbl=MC_even_dist.summarize_cumulative_return()
# Print summary statistics
print(even_tbl)
```

```python
# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $15,000 investments stocks
initial_inv=20000
# Print results
print(f"There is a 95% chance that an initial investment of ${initial_inv} in the portfolio over the next 5 years will end within in the range of {round(even_tbl[8]*initial_inv,2)} and {round(even_tbl[9]*initial_inv,2)}.")
```

```python
#Calculate the expected portfolio return at the 95% lower and upper confidence intervals based on a 50% increase in the initial investment.
New_inv=initial_inv * 1.5
print(f"There is a 95% chance that an initial investment of ${New_inv} in the portfolio over the next 5 years will end within in the range of {round(even_tbl[8]*New_inv,2)} and {round(even_tbl[9]*New_inv,2)}.")
```
